```python
from bs4 import BeautifulSoup
import requests
import shutil
import matplotlib.pyplot as plt
import matplotlib.image as mpim
```

## Scraping from https://unsplash.com/s/photos/abstract-art

```python
html_page = requests.get('https://unsplash.com/s/photos/abstract-art')
soup = BeautifulSoup(html_page.content, 'html.parser')
warning = soup.find('div', class_="_1ZjfQ")
book_container = warning.nextSibling
```

```python
images = book_container.findAll('img')
example = images[1]
example
```

```python
images
```

```python
example.attrs['src']
```

```python
url = example.attrs['src']
r = requests.get(url, stream=True)
if r.status_code == 200:    
   with open("book1.jpg", 'wb') as f: 
      r.raw.decode_content = True
      shutil.copyfileobj(r.raw, f)
```

```python
img = mpim.imread('book1.jpg')
imgplot = plt.imshow(img)
plt.show()
```
